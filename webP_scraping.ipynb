```python
final=pd.DataFrame()
for j in range(1,11):
    URL = 'https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget={}'.format(j)
    HEADERS = ({'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:124.0) Gecko/20100101 Firefox/124.0', 'Accpet-Language': 'en-US, en;q=0.5'})
    webpage= requests.get(URL,headers=HEADERS)
    soup = BeautifulSoup(webpage.content, 'html.parser')
    company = soup.find_all('div', class_='companyCardWrapper')
    name=[]
    rating=[]
    reviews=[]
    ctype=[]
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_='companyCardWrapper__companyRating').text.strip())
        reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip())
        ctype.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.strip())
        
    d={'name': name, 'rating': rating, 'reviews': reviews, 'Company Details': ctype}
    df=pd.DataFrame(d)
    final.append(df)
```
